```python
import os
import pandas as pd
```

```python
csv_names = os.listdir()  # get names of all files under this directory
```

```python
csv_names = list(filter(lambda s: '.csv' in s and '2021' in s, csv_names))  # filter out csv files for 2021
```

```python
# devide names w.r.p to their months
DF = []  # stores the dataframe derived from the csv files
for name in csv_names:
    month = name[:2]
    date = name[3: 5]
    year = '2021'
    df = pd.read_csv(name)
    df['month'] = month
    df['date'] = date
    df['year'] = year
    DF.append(df)
```

```python
covid_dataset = pd.concat(DF)  # concate (union) all dataframes
```

```python
covid_dataset.to_csv('daily_covid-19_cases_in_the_us_in_2021.csv')
```
