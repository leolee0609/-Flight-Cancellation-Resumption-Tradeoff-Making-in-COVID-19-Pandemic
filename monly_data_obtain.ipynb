```python
import os
import pandas as pd
```

```python
df = pd.read_csv('Daily_Covod19_US.csv')
```

```python
df
```

```python
states = set(df['Province_State'])
```

```python
max_date = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
```

```python
Rows = []
for state in states:
    for month in range(1, 11):

        row_in_df_1 = df.loc[df['Province_State'] == state].loc[df['month'] == month].loc[df['date'] == 1]
        row_in_df_31 = df.loc[df['Province_State'] == state].loc[df['month'] == month].loc[df['date'] == max_date[month]]
        
        if row_in_df_31.empty or row_in_df_1.empty:
            continue
            
        Confirmed_overall = (row_in_df_31['Confirmed'].reset_index()['Confirmed'][0] + row_in_df_1['Confirmed'].reset_index()['Confirmed'][0]) / 2
        
        Test_overall = (row_in_df_31['Total_Test_Results'].reset_index()['Total_Test_Results'][0] + row_in_df_1['Total_Test_Results'].reset_index()['Total_Test_Results'][0]) / 2
        
        Increased_Confirmed = row_in_df_31['Confirmed'].reset_index()['Confirmed'][0] - row_in_df_1['Confirmed'].reset_index()['Confirmed'][0]       
        
            
        Increased_Deaths = row_in_df_31['Deaths'].reset_index()['Deaths'][0] - row_in_df_1['Deaths'].reset_index()['Deaths'][0]
          
            
        Increased_Recovered = row_in_df_31['Recovered'].reset_index()['Recovered'][0] - row_in_df_1['Recovered'].reset_index()['Recovered'][0]
                       
        
            
        Total_test_in_the_month = row_in_df_31['Total_Test_Results'].reset_index()['Total_Test_Results'][0] - row_in_df_1['Total_Test_Results'].reset_index()['Total_Test_Results'][0]
        
        
        
        
        Rows.append((state, month, Increased_Confirmed, Increased_Deaths, Increased_Recovered, Total_test_in_the_month, Confirmed_overall, Test_overall))
```

```python
monthly_data = pd.DataFrame(data = Rows, columns = ['State','month', 'Increased_Confirmed', 'Increased_Deaths', 'Increased_Recovered', 'Total_test_in_the_month', 'Confirmed_overall', 'Test_overall'])
```

```python
monthly_data.to_csv('monthly_covid_increase_data_by_state_2021.csv')
```
